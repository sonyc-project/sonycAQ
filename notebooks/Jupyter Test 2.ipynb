## Setup Code <a class="anchor" id="Header-1"></a>

This code initializes pandas and imports the data from the .csv file.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv (r'C:\Users\vendi\Desktop\Internships and Work\NYU\Test Data\dusty_book_test.csv')
x = data['datetime']
y1 = data['PC0.1']
y2 = data['PC0.3']
y3 = data['PC0.5']
y4 = data['PC1.0']
y5 = data['PC2.5']
y6 = data['PC5.0']
y7 = data['PC10']
y8 = data['PM0.1']
y9 = data['PM0.3']
y10 = data['PM0.5']
y11 = data['PM1.0']
y12 = data['PM2.5']
y13 = data['PM5.0']
y14 = data['PM10']
```

## PC Plots <a class="anchor" id="Header-2"></a>

This code creates each of the particle count plots.

```python
plt.plot(x,y1)
plt.xlabel('Time (s)')
plt.ylabel('PC (#/L)')
plt.title('PC0.1')
plt.show()

plt.plot(x,y2)
plt.xlabel('Time (s)')
plt.ylabel('PC (#/L)')
plt.title('PC0.3')
plt.show()

plt.plot(x,y3)
plt.xlabel('Time (s)')
plt.ylabel('PC (#/L)')
plt.title('PC0.5')
plt.show()

plt.plot(x,y4)
plt.xlabel('Time (s)')
plt.ylabel('PC (#/L)')
plt.title('PC1.0')
plt.show()

plt.plot(x,y5)
plt.xlabel('Time (s)')
plt.ylabel('PC (#/L)')
plt.title('PC2.5')
plt.show()

plt.plot(x,y6)
plt.xlabel('Time (s)')
plt.ylabel('PC (#/L)')
plt.title('PC5.0')
plt.show()

plt.plot(x,y7)
plt.xlabel('Time (s)')
plt.ylabel('PC (#/L)')
plt.title('PC10')
plt.show()
```

## PM Plots <a class="anchor" id="Header-3"></a>

This code creates each of the particle mass plots.

```python
plt.plot(x,y8)
plt.xlabel('Time (s)')
plt.ylabel('PM (ug/m^3)')
plt.title('PM0.1')
plt.show()

plt.plot(x,y9)
plt.xlabel('Time (s)')
plt.ylabel('PM (ug/m^3)')
plt.title('PM0.3')
plt.show()

plt.plot(x,y10)
plt.xlabel('Time (s)')
plt.ylabel('PM (ug/m^3)')
plt.title('PM0.5')
plt.show()

plt.plot(x,y11)
plt.xlabel('Time (s)')
plt.ylabel('PM (ug/m^3)')
plt.title('PM1.0')
plt.show()

plt.plot(x,y12)
plt.xlabel('Time (s)')
plt.ylabel('PM (ug/m^3)')
plt.title('PM2.5')
plt.show()

plt.plot(x,y13)
plt.xlabel('Time (s)')
plt.ylabel('PM (ug/m^3)')
plt.title('PM5.0')
plt.show()

plt.plot(x,y14)
plt.xlabel('Time (s)')
plt.ylabel('PM (ug/m^3)')
plt.title('PM10')
plt.show()
```

## Analysis <a class="anchor" id="Header-4"></a>

We can observe that the behavior of the particles is how we might expect - PC and PM spike when the book is slammed and decrease when the book is removed. PM is higher for larger masses even though there are fewer particles present. Particle counts spike at the same time as one another. The table of contents works properly.
