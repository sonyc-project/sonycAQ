# Test Using `for` Loop to Import All Files

This notebook is testing the use of `for` loops for all functions for scaleability and shorter code.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Construct Data Set

Import filenames

```python
directory = os.listdir(r'C:\Users\vendi\Desktop\Internships and Work\NYU\Minute-by-Minute Tests')
```

```python
j = 0
data = [''] * len(directory)
for i in directory:
    path = r'C:\Users\vendi\Desktop\Internships and Work\NYU\Minute-by-Minute Tests' + "\\" + i
    data[j] = pd.read_csv(path)
    j = j + 1
```

## Put Data From Different Files Together

```python
check = 0                                       # can't append first variable, need this variable to initialize x
for i in data:
    if check == 0:
        x = i['datetime']
        check = 1
    else:
        x = x.append(i['datetime'])
```

## Print Data

```python
for j in data[1]:                               # iterate through each piece of data (all PC's, all PM's)
    check = 0                                   # can't append first variable, need this variable to initialize y
    for i in data:                              # append data from all csv's together
        if check == 0:
            y = i[j]
            check = 1
        else:
            y = y.append(i[j])
    if j != 'datetime':                         # don't want to plot time vs. time!
        plt.plot(x,y)
        plt.xlabel('Time (s)')
        if j[1] == 'C':                         # need appropriate units depending on PC or PM
            plt.ylabel('#/L')
        else:
            plt.ylabel('ug/m^3')
        plt.title(j)
        plt.show()
```
