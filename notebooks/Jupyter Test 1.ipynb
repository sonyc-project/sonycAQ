# Test 2: PC from Jun. 25, 2021 - 21:26 (with subplots)

Test drawing all PC data from a single .csv at once using the subplot function.

```python
data = pd.read_csv (r'C:\Users\vendi\Desktop\Internships and Work\NYU\Minute-by-Minute Tests\test_node-venditto-pm-2021_06_25-21_26_08.csv')
x = data['datetime']
y1 = data['PC0.1']
y2 = data['PC0.3']
y3 = data['PC0.5']
y4 = data['PC1.0']
y5 = data['PC2.5']
y6 = data['PC5.0']
y7 = data['PC10']
fig = plt.figure()

P = fig.add_subplot(111)
p1 = fig.add_subplot(421)
p2 = fig.add_subplot(422)
p3 = fig.add_subplot(423)
p4 = fig.add_subplot(424)
p5 = fig.add_subplot(425)
p6 = fig.add_subplot(426)
p7 = fig.add_subplot(427)

p1.plot(x,y1)
p2.plot(x,y2)
p3.plot(x,y3)
p4.plot(x,y4)
p5.plot(x,y5)
p6.plot(x,y6)
p7.plot(x,y7)

P.set_xlabel('Time(s)')
P.set_ylabel('PC (#/L)')
P.set_title('PC')

p1.set_title('PC0.1')
p2.set_title('PC0.3')
p3.set_title('PC0.5')
p4.set_title('PC1.0')
p5.set_title('PC2.5')
p6.set_title('PC5.0')
p7.set_title('PC10')


plt.show()
```
